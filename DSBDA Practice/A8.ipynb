 code to check how the price of the ticket (column name: 'fare') for each passenger is distributed by plotting a histogram. 

```python
sns.histplot(dataset['fare'], kde=True, linewidth=0);
```

```python
sns.jointplot(x='age', y='fare', data=dataset);
```
